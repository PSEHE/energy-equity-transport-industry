```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
#import geopandas as gp
```

#### Note: Need to redownload ACS commute data from NHGIS to run.

```python
commute = pd.read_csv('../Demographic/2018acs5yr/nhgis0009_commute_csv/nhgis0009_ds240_20185_2018_tract.csv', engine='python')
```

```python
mapping = {'AKAPE001':    'Total',
        'AKAPE002':    'Below 100 percent of the poverty level',
        'AKAPE003':    '100 to 149 percent of the poverty level',
        'AKAPE004':    'At or above 150 percent of the poverty level',
        'AKAPE005':    'Car, truck, or van - drove alone',
        'AKAPE006':    'Car, truck, or van - drove alone: Below 100 percent of the poverty level',
        'AKAPE007':    'Car, truck, or van - drove alone: 100 to 149 percent of the poverty level',
        'AKAPE008':    'Car, truck, or van - drove alone: At or above 150 percent of the poverty level',
        'AKAPE009':    'Car, truck, or van - carpooled',
        'AKAPE010':    'Car, truck, or van - carpooled: Below 100 percent of the poverty level',
        'AKAPE011':    'Car, truck, or van - carpooled: 100 to 149 percent of the poverty level',
        'AKAPE012':    'Car, truck, or van - carpooled: At or above 150 percent of the poverty level',
        'AKAPE013':    'Public transportation (excluding taxicab)',
        'AKAPE014':    'Public transportation (excluding taxicab): Below 100 percent of the poverty level',
        'AKAPE015':    'Public transportation (excluding taxicab): 100 to 149 percent of the poverty level',
        'AKAPE016':    'Public transportation (excluding taxicab): At or above 150 percent of the poverty level',
        'AKAPE017':    'Walked',
        'AKAPE018':   'Walked: Below 100 percent of the poverty level',
        'AKAPE019':    'Walked: 100 to 149 percent of the poverty level',
        'AKAPE020':    'Walked: At or above 150 percent of the poverty level',
        'AKAPE021':    'Taxicab, motorcycle, bicycle, or other means',
        'AKAPE022':    'Taxicab, motorcycle, bicycle, or other means: Below 100 percent of the poverty level',
        'AKAPE023':    'Taxicab, motorcycle, bicycle, or other means: 100 to 149 percent of the poverty level',
        'AKAPE024':    'Taxicab, motorcycle, bicycle, or other means: At or above 150 percent of the poverty level',
        'AKAPE025':    'Worked at home',
        'AKAPE026':    'Worked at home: Below 100 percent of the poverty level',
        'AKAPE027':    'Worked at home: 100 to 149 percent of the poverty level',
        'AKAPE028':    'Worked at home: At or above 150 percent of the poverty level'}
```

```python
for key, value in mapping.items():
    commute.columns = [column.replace(key, value) for column in commute]
```

```python
commute.drop(columns = commute.columns[commute.columns.str.contains('AKAPM')], inplace=True)
```

```python
commute['GISJOIN'] = commute['GISJOIN'].str[1:]
commute_nvnmco = commute.loc[(commute['STATE'] == 'Colorado') | (commute['STATE'] == 'New Mexico') | (commute['STATE'] == 'Nevada')]
commute_nvnmco['GISJOIN'] = commute_nvnmco['GISJOIN'].str[0:2:] + commute_nvnmco['GISJOIN'].str[2 + 1::]
commute_nvnmco['GISJOIN'] = commute_nvnmco['GISJOIN'].str[0:5:] + commute_nvnmco['GISJOIN'].str[5 + 1::]
commute_nvnmco['GISJOIN'] = commute_nvnmco['GISJOIN'].str.lstrip('0')

commute_nvnmco.to_csv('Demographic/2018acs5yr/ACS5Y2018_Commute.csv')
```
